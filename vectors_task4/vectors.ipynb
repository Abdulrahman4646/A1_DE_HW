```python
def text_to_vector(text, bag_of_words):

    words = text.lower().split()
    
    vector = []
    
    for word in bag_of_words:
        if word in words:
            vector.append(words.count(word))
        else:
            vector.append(0)
    
    return vector


bag_of_words = ["i", "love", "python", "data"]

sentence = "I love python"

vector = text_to_vector(sentence, bag_of_words)

print(vector)
```
