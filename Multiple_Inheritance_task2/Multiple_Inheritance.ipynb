# الوراثة المتعددة (Multiple Inheritance) في بايثون

الوراثة المتعددة تعني أن الكلاس (Class) يمكن أن يرث من أكثر من كلاس في نفس الوقت.

بمعنى آخر:
الكلاس الجديد يستطيع أن يأخذ الخصائص (المتغيرات) والدوال (Functions) من أكثر من كلاس أب.

لماذا نستخدمها؟
نستخدم الوراثة المتعددة عندما نريد دمج صفات أو وظائف من أكثر من كلاس في كلاس واحد.

مثال بسيط:
لو عندنا كلاس يمثل "أب" وكلاس يمثل "أم"،
نستطيع إنشاء كلاس "ابن" يرث من الاثنين معاً ويأخذ الصفات من كليهما.

الفائدة:
تساعدنا على إعادة استخدام الكود وتنظيمه بشكل أفضل.

```python
class Father:
    def skill_father(self):
        print("I can fix things.")


class Mother:
    def skill_mother(self):
        print("I can cook delicious food.")


class Child(Father, Mother):
    pass


child1 = Child()

child1.skill_father()
child1.skill_mother()
```
