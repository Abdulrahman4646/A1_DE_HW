ملخص عام عن البيانات

```python
df.describe()
```

حذف الصفوف المكررة إذا كانت موجودة

```python
df1 = df.drop_duplicates()
```

التعامل مع القيم الفارغة

حذف القيم الفارغة التي في عمود ال item كونها لا يمكن تعبئتها

```python
df1.dropna(subset=["Item"], inplace = True)
```

```python
df1.isna().sum()
```

تعبئة القيم الفارغة في عمود سعر الواحدة بالمتوسط

```python
df1["Price Per Unit"] = df1["Price Per Unit"].fillna(df1["Price Per Unit"].mean())
```

```python
df1.isna().sum()
```

تعبئة القيم الفارغة في عمود الإجمالي بقيمة الضرب بين عمود الكمية وعمود سعر الواحدة

```python
df1["Total Spent"] = df1["Price Per Unit"] * df1["Quantity"]
```

```python
df1.isna().sum()
```

تعبئة القيم الفارغة في عمود المنطقة بقيمة unknown

```python
df1["Location"].fillna("Unknown", inplace = True)
```

```python
df1.isna().sum()
```

تعبئة القيم الفارغة في عمود الخصم بقيمة false

```python
df1["Discount Applied"].fillna("False", inplace = True)
```

```python
df1.isna().sum()
```

نتأكد من القيم داخل كل عمود

```python
df1['Category'].value_counts()
```

```python
df1['Payment Method'].value_counts()
```

```python
df1['Location'].value_counts()
```

تحويل نوع بيانات التاريخ

```python
df["Transaction Date"] = pd.to_datetime(df["Transaction Date"])
```

حفظ البيانات النظيفة

```python
df.to_csv("retail_store_sales_cleaned.csv", index=False)
```
