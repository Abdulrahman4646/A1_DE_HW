# الحزم (Packages) والموديولات (Modules) الجاهزة في بايثون

في بايثون في أشياء جاهزة نقدر نستخدمها بدون ما نكتبها من الصفر.
هذه الأشياء تنقسم إلى:

1) Module (موديول):
هو ملف بايثون فيه دوال أو أدوات جاهزة نقدر نستوردها ونستخدمها.
مثال: موديول الرياضيات math.

2) Package (حزمة):
هي مجموعة موديولات موجودة داخل مجلد واحد.
يعني الحزمة تحتوي على أكثر من موديول منظمين مع بعض.

ليش نستخدمهم؟
عشان يوفروا علينا وقت، بدل ما نكتب كل شيء من البداية.
بايثون فيه مكتبة ضخمة جاهزة اسمها Standard Library فيها موديولات كثيرة مفيدة.

أمثلة على موديولات جاهزة:
- math → للعمليات الرياضية
- random → للأرقام العشوائية
- datetime → للتاريخ والوقت
- os → للتعامل مع النظام والملفات

باختصار:
الموديول = ملف فيه أدوات جاهزة
الباكيج = مجلد فيه أكثر من موديول

```python
import math
import random
import datetime

print(math.sqrt(16))

print("\n", random.randint(1, 10))

today = datetime.date.today()
print("\n", today)
```
